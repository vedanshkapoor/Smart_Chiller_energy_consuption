```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/DAV Project/HVAC Energy Data (1) - HVAC Energy Data (1).csv')

columns_to_drop = ['Outside Temperature (F)', 'Dew Point (F)', 'Wind Speed (mph)', 'Pressure (in)']
df = df.drop(labels=columns_to_drop, axis=1)

corr_drop = ['Local Time (Timezone : GMT+8h)', 'Date','Time']
df_corr_matrix = df.drop(labels = corr_drop, axis = 1 )
corr_matrix = df_corr_matrix.corr()

sns.heatmap(corr_matrix, annot = True , cmap = 'coolwarm')
```

```python

plt.figure(figsize=(10, 6))
x_values = pd.to_datetime(df['Local Time (Timezone : GMT+8h)'])
filtered_indices = x_values < pd.to_datetime('8/20/2019 0:00:00')
plt.plot(x_values[filtered_indices],df['Chiller Energy Consumption (kWh)'][filtered_indices])
plt.xlabel('Local Time')
plt.ylabel('Chiller Energy Consumption (kWh)')
plt.title('Chiller Energy Consumption Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
